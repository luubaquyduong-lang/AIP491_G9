```python
file_path = '/content/drive/MyDrive/KY_9/KY_9_Final_Project/Data/processed/data_final_train_v1/data_final_sort_v1.jsonl'
```

```python
import json
import pickle
from sentence_transformers import SentenceTransformer
```

```python
passages = []
with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        try:
            obj = json.loads(line)
            passages.append(obj["passage"])
        except json.JSONDecodeError:
            print("⚠️ Lỗi JSON ở dòng:", len(passages))
            continue

print("✅ Đọc được:", len(passages))
```

**bkai_foundation_models_fine_tuning**

```python
model_path_bkai = '/content/drive/MyDrive/KY_9/KY_9_Final_Project/Data/train_model/bkai_foundation_models_fine_tuning'
model_bkai = SentenceTransformer(model_path_bkai)
#
output_path_bkai = '/content/drive/MyDrive/KY_9/KY_9_Final_Project/Data/embeddings/model_fine_tuning/embeding_by_model_fine_bkai_v1.pkl'
# ✅ Tạo thư mục nếu chưa tồn tại
os.makedirs(os.path.dirname(output_path_bkai), exist_ok=True)
```

```python
# Tạo embedding (batch encode)
embeddings = model_bkai .encode(passages, show_progress_bar=True)
```

```python
import pickle
import os
from IPython.display import FileLink

os.makedirs(os.path.dirname(model_path_bkai), exist_ok=True)
with open(output_path_bkai, "wb") as f:
    pickle.dump(embeddings, f)

print("✅ Đã lưu:", output_path_bkai)
FileLink(output_path_bkai)
```
